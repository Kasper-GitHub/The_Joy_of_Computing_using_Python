# Q-3

```python
import networkx as nx
#import matplotlib.pyplot as plt

True_Count = 0
False_Count = 0

for i in range(100):
    G = nx.gnp_random_graph(10,0.5)
    #nx.draw(G)
    #plt.show()
    #print(nx.is_connected(G))
    Count = nx.is_connected(G)
    if Count == True:   True_Count = True_Count + 1
    else: False_Count = False_Count + 1

print("True : ", True_Count,"\tFalse : ",False_Count)
```

# Q-4

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (2, 1), (2, 3), (3,4), (4,1), (3,1)])
nx.draw(G, with_labels=True)
plt.show()
```
